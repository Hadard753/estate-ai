```python
import requests
import pandas
import xml.etree.ElementTree as et 

def readPageData(objectID, PageNo):
    requestBody = { "MoreAssestsType": 0, "FillterRoomNum": 0, "GridDisplayType": 0, "ResultType": 1, "ObjectID": objectID, "ObjectIDType": "text", "ObjectKey": "UNIQ_ID", "DescLayerID": "SETL_MID_POINT", "Alert": None, "Gush": "", "Parcel": "", "showLotParcel": False, "showLotAddress": False, "MutipuleResults": False, "ResultsOptions": None, "CurrentLavel": 2, "isHistorical": False, "PageNo": PageNo, "OrderByFilled": "DEALDATETIME", "OrderByDescending": True, "Distance": 0 };

    result = requests.post("https://www.nadlan.gov.il/Nadlan.REST/Main/GetAssestAndDeals", requestBody);
    
    print("Page "+str(PageNo) +" loaded "+str(len(result.json()["AllResults"]))+" , last: "+str(result.json()["IsLastPage"]));
    if not result.json()["IsLastPage"]:
        PageNo += 1;
        df = pandas.DataFrame(result.json()["AllResults"]);
        df.to_csv("./sales.csv", mode='a', header=False);
        readPageData(objectID, PageNo);
        
def mapCity(c) :
    return {"name_he": c["שם_ישוב"], "name_en": c["שם_ישוב_לועזי"]};

        
def getListOfCities(cities, url):
    response = requests.get("https://data.gov.il"+url);
    print("https://data.gov.il"+url);
    data = response.json()["result"];
    result = data["records"];
    cities.extend(list(map(mapCity, result)));
    
    next_url = data["_links"]["next"];
    total = data["total"];
    if not "offset" in data or data["offset"]<total:
        getListOfCities(cities, next_url)

def getObjectID(city):
    #params = querystring.stringify({query: city.name_he});
    response = requests.get("https://www.nadlan.gov.il/Nadlan.REST/Main/GetDataByQuery", params={"query": city["name_he"]});
    return response.json()["ObjectID"];

cities = [];
getListOfCities(cities, "/api/3/action/datastore_search?resource_id=5c78e9fa-c2e2-4771-93ff-7f400a12f7ba&include_total=true");
print("Found "+str(len(cities))+" Cities");

df = pandas.DataFrame(cities);
df.to_csv("./cities.csv");

for i in cities:
    objectID = getObjectID(i);
    readPageData(objectID, 1);
```
